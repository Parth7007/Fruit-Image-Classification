```python
from tensorflow import keras 
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Activation, Dropout, Flatten, Dense
from keras.preprocessing.image import  img_to_array, load_img 
```

```python
train_path = 'fruits-360_dataset_100x100/fruits-360/Training/'
test_path = 'fruits-360_dataset_100x100/fruits-360/Test/'
```

```python
img = load_img(train_path + 'Orange/0_100.jpg')
plt.imshow(img)
plt.axis("on")
plt.show()
```

```python
img = img_to_array(img)
print(img)
```

```python
img.shape
```

```python
#building the model
model = Sequential()
model.add(Conv2D(128, 3, activation="relu", input_shape = (100, 100, 3)))
model.add(MaxPooling2D())
model.add(Conv2D(64, 3, activation = "relu"))
model.add(Conv2D(32, 3, activation = "relu"))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(10000, activation = "relu"))
model.add(Dense(5000, activation = "relu"))
model.add(Dense(1000, activation = "relu"))
model.add(Dense(141, activation = "softmax"))
model.summary()
```

```python
#Compiling The Model
model.compile(loss="categorical_crossentropy", optimizer = "SGD", metrics = ["accuracy"])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                  shear_range = 0.3,
                  horizontal_flip=True,
                  vertical_flip=False,
                  zoom_range = 0.3
                  )
test_datagen  = ImageDataGenerator(rescale = 1./255)

train_generator = train_datagen.flow_from_directory(train_path,
                                                    target_size=(100,100),
                                                    batch_size = 32,
                                                    color_mode= "rgb",
                                                    class_mode = "categorical")
test_generator = test_datagen.flow_from_directory(test_path,
                                                    target_size=(100,100),
                                                    batch_size = 32,
                                                    color_mode= "rgb",
                                                    class_mode = "categorical")
```

```python
hist = model.fit(
    train_generator,                # Pass the training data generator directly
    steps_per_epoch=50,             # Number of steps per epoch
    epochs=50,                      # Number of epochs
    validation_data=test_generator, # Validation data generator
    validation_steps=50            # Number of validation steps
)
```

```python
from tensorflow.keras.models import load_model

# Load the trained model
model = load_model('FruitModel.h5')
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np

# Load and preprocess the image
def preprocess_image(image_path):
    img = load_img(image_path, target_size=(image_width, image_height))  # Resize image
    img_array = img_to_array(img)  # Convert to array
    img_array = np.expand_dims(img_array, axis=0)  # Expand dimensions to match model's input shape
    img_array /= 255.0  # Normalize the image
    return img_array

image_path = 'fruits-360_dataset_100x100/fruits-360/Test/Fig/27_100.jpg'
image_width, image_height = 100, 100  # Set to the input size used during training
image = preprocess_image(image_path)
```

```python
# Predict the class of the fruit
predictions = model.predict(image)
predicted_class_index = np.argmax(predictions, axis=1)  # Get the index of the highest probability

print(f"Predicted class index: {predicted_class_index[0]}")
```

```python
import matplotlib.pyplot as plt

plt.imshow(load_img(image_path))
plt.title(f"Predicted class index: {predicted_class_index[0]}")
plt.axis('off')
plt.show()
```
